### Loading the black and white image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('Data/zidane_gray.jpg' , cv2.IMREAD_GRAYSCALE)  # Read image as grayscale


plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

### Disply the reults of edge detected image with original image

```python
from skimage import filters
from skimage import io, color

def display_results(original, image, image_type):
    plt.figure(figsize=(10, 8))
    plt.subplot(2, 2, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Original Image')
    
    plt.subplot(2, 2, 2)
    plt.imshow(image, cmap='gray')
    plt.title(image_type + ' Edge Detection')

    plt.tight_layout()
    plt.show()
```

### Sobel Edge Detection

```python
def sobel_edge_detection(image):
    sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Sobel along X-axis
    sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Sobel along Y-axis
    sobel = cv2.magnitude(sobel_x, sobel_y)  # Magnitude of the gradient
    return sobel

sobel_edges = sobel_edge_detection(image)

display_results(image, sobel_edges, 'Sobel')
```

### Prewitt Edge Detection

```python
def prewitt_edge_detection(image):
    prewitt_x = filters.prewitt_h(image)  # Prewitt along X-axis
    prewitt_y = filters.prewitt_v(image)  # Prewitt along Y-axis
    prewitt = np.sqrt(prewitt_x**2 + prewitt_y**2)  # Magnitude of the gradient
    return prewitt

prewitt_edges = prewitt_edge_detection(image)

display_results(image, prewitt_edges, 'Prewitt')
```

### Roberts Edge Detection

```python
def roberts_edge_detection(image):
    roberts = filters.roberts(image)  # Roberts edge detection
    return roberts

roberts_edges = roberts_edge_detection(image)

display_results(image, roberts_edges, 'Roberts')
```
