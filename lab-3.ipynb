### Generating 35 random integers between 10 and 100

```python
import numpy as np

data = np.random.randint(10, 101, 40)
print("Data Points:",data)
```

### Defining bins

```python
bins = list(range(10, 110, 10))
```

### Categorizing data into bins and counting frequencies

```python
import pandas as pd

frequency_counts = pd.cut(data, bins=bins, right=False).value_counts().sort_index()
```

### Creating a table of frequency counts

```python
frequency_table = pd.DataFrame(frequency_counts).reset_index()
frequency_table.columns = ['Range', 'Frequency']
print(frequency_table)
```

### Plotting the data

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(frequency_table['Range'].astype(str), frequency_table['Frequency'], marker='o', linestyle='-')
plt.xlabel('Range')
plt.ylabel('Frequency')
plt.title('Frequency of Data Points in Ranges')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
