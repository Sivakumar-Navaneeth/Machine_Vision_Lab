### Loading the image with black and white noise

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('Data/noise_image.png')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  

plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

### Add Median Filter

```python
median_filtered = cv2.medianBlur(image, 5)

median_filtered_rgb = cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 8))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(median_filtered_rgb)
plt.title('Median Filter')
plt.axis('off')

plt.show()
```

### Add Minimum Filter

```python
kernel = np.ones((5,5), np.uint8)
min_filtered = cv2.erode(image, kernel)

min_filtered_rgb = cv2.cvtColor(min_filtered, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(10, 8))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')


plt.subplot(1, 2, 2)
plt.imshow(min_filtered_rgb)
plt.title('Minimum Filter')
plt.axis('off')

plt.show()
```

### Add Maximum Filter

```python
max_filtered = cv2.dilate(image, kernel)

max_filtered_rgb = cv2.cvtColor(max_filtered, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 8))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')


plt.subplot(1, 2, 2)
plt.imshow(max_filtered_rgb)
plt.title('Maximum Filter')
plt.axis('off')

plt.show()
```
