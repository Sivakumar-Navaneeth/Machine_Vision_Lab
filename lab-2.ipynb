### 1. Read Image in Grayscale

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('Data/zidane.jpg', cv2.IMREAD_GRAYSCALE)

plt.title('Original Image')
plt.imshow(image, cmap='gray')
```

### 2. Plot Image Negative

```python
image_negative = 255 - image

plt.title('Image Negative')
plt.imshow(image_negative, cmap='gray')
```

### 3. Plot Gamma corrected Image

```python
gamma = 2.2
image_gamma = np.array(255 * (image / 255) ** gamma, dtype='uint8')

plt.title('Gamma Correction')
plt.imshow(image_gamma, cmap='gray')
```

### 4. Plot Log Transformatted Image

```python
c = 255 / np.log(1 + np.max(image))
image_log = c * (np.log(1 + image))
image_log = np.array(image_log, dtype='uint8')

plt.title('Log Transformation')
plt.imshow(image_log, cmap='gray')
```

### 5. Plot Low Pass Filter Image

```python
kernel = np.ones((5, 5), np.float32) / 25
image_low_pass = cv2.filter2D(image, -1, kernel)

plt.title('Low Pass Filter')
plt.imshow(image_low_pass, cmap='gray')
```

### 6. Plot High Pass Filter Image

```python
kernel = np.array([[-1, -1, -1], 
                   [-1,  8, -1], 
                   [-1, -1, -1]])
image_high_pass = cv2.filter2D(image, -1, kernel)

plt.title('High Pass Filter')
plt.imshow(image_high_pass, cmap='gray')
```

### 7. Plot Gaussian Filter Image

```python
image_gaussian = cv2.GaussianBlur(image, (5, 5), 0)

plt.title('Gaussian Filter')
plt.imshow(image_gaussian, cmap='gray')
```

### 8. Plot Median Filter Image

```python
image_median = cv2.medianBlur(image, 5)

plt.title('Median Filter')
plt.imshow(image_median, cmap='gray')
```
